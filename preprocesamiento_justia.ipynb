```python
import re
import json
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# 1. Cargar corpus simulado
with open('corpus_original.txt', 'r', encoding='utf-8') as f:
    textos = f.readlines()

# 2. Función de limpieza
def limpiar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r'[^\w\s]', '', texto)  # elimina puntuación
    texto = re.sub(r'\d+', '', texto)      # elimina números
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

# 3. Tokenización y stopwords (español)
stop_words = set(stopwords.words('spanish'))
lemmatizer = WordNetLemmatizer()  # Requiere descargar wordnet en español

def preprocesar(texto):
    texto = limpiar_texto(texto)
    tokens = word_tokenize(texto, language='spanish')
    tokens = [t for t in tokens if t not in stop_words]
    tokens = [lemmatizer.lemmatize(t) for t in tokens]  # simplificado
    return ' '.join(tokens)

corpus_limpio = [preprocesar(t) for t in textos]

# Guardar
with open('corpus_limpio.json', 'w', encoding='utf-8') as f:
    json.dump(corpus_limpio, f, indent=2)
```
